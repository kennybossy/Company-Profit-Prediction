```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('C:\\Users\\ACER\Downloads\\50_Startups.csv')
x = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
y_predict=model.predict(x_train)
y_predict 
```

```python
m=model.coef_
c=model.intercept_
print(f'c={c}, m={m}')
```

```python
from sklearn.metrics import r2_score 
r2_score(y_train,y_predict)
y_predict_test=model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
r2_1=r2_score(y_train,y_predict)
mse_1=mean_squared_error(y_train,y_predict)
mae_1=mean_absolute_error(y_train,y_predict)
print("r2_score: ",r2_1)
print("mean_squared_error: ",mse_1)
print("mean_absolute_error: ",mae_1)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
r2_2=r2_score(y_test,y_predict_test)
mse_2=mean_squared_error(y_test,y_predict_test)
mae_2=mean_absolute_error(y_test,y_predict_test)
print("r2_score: ",r2_2)
print("mean_squared_error: ",mse_2)
print("mean_absolute_error: ",mae_2)
```

```python
plt.xlabel("x_train")
plt.ylabel("y_train")
plt.scatter(x_train,y_train)
plt.plot(x_train,y_predict,color='r')
plt.show()
```

```python
plt.xlabel("x_test")
plt.ylabel("y_test")
plt.scatter(x_test,y_test)
plt.plot(x_train,y_predict,color='r')
plt.show()
```
